Type Conversion


```python
df['contact_tracing'] = df['contact_tracing'].astype(int)
df['emergency_investment_in_healthcare'] = df['emergency_investment_in_healthcare'].astype(int)
```

```python
df['Total_Policy'] = df['testing_policy'] + df['vaccination_policy']  # create a new columns
```

**Used Matplotlib**

Draw a line chart with the help of the vaccination_policy and testing_policy

```python
plt.plot(df['vaccination_policy'].head(700), df['testing_policy'].head(700), marker='o', color="blue", linestyle="solid", linewidth=1)
plt.xlabel('Vaccination Policy')
plt.ylabel('Testing Policy')
plt.title('Policy')
plt.grid()
plt.show()
```

Draw bar chart

```python
plt.bar(df['location_key'].head(6000), df['investment_in_vaccines'].head(6000), color="grey", width=0.5, label="Vaccines 2025")
plt.xlabel('Location Key')
plt.ylabel('Investment in Vaccines')
plt.title('Vaccines')
plt.legend()
plt.tight_layout()
plt.show()
```

Histrogram

```python
plt.hist(df['income_support'], bins=6)
plt.xlabel('income')
plt.ylabel('Frequency')
plt.title('Income')
plt.show()
```

Scatterplot

```python
plt.scatter(df['School Closing'].head(400), df['workplace_closing'].head(400), color="Red", marker="o", label="Scatter Plot")
plt.xlabel('school')
plt.ylabel('workplace')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```
